```python
import requests

base_url = "http://localhost:9000/"
```

#### Test Connection

```python
response = requests.get(base_url)
response.json()
```

#### Create a session

```python
url = base_url + "new_session"
response = requests.post(url)

session_id = response.json()['session_id']
response.json()
```

```python
type(session_id)
```

#### Request Chat Completion

```python
url = base_url + f"chat"
message = "Give me papers by Protopapas"
response = requests.post(url, json={"session_id":session_id, "message": message})
response.json()
```

#### Request Session History

```python
url = base_url + f"sessions/{session_id}"
response = requests.get(url, json={"session_id":session_id})
response.json()
```

#### Reset Memory

```python
url = base_url + f"reset/{session_id}"
response = requests.get(url, json={"session_id":session_id})
response.json()
```
